#### 1 - Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json

```python
import requests
from pprint import pprint
```

```python
url = 'https://api.github.com/users/SpaceMakerc/repos'
```

```python
response = requests.get(url = url)
print(response.status_code)
```

```python
pprint(response.json())
```

```python
lists = response.json()
```

```python
with open('list_repo.json', 'w') as repos:
    for i in lists:
        print(i['name'])
        repos.write(i['name'] + '\n')
```

#### 2 - Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search
#### (а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа.
#### (б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

```python
params = {'latitude' : '56.012073', 'longitude' : '92.863298', 'query' : 'шашлык'}
url = 'https://api.delivery-club.ru/api1.2/vendors/search'
```

```python
response = requests.get(url = url, params = params)
print(respons.status_code)
```

```python
pprint(response.json())
```

```python
vendors = response.json()['vendors']
all_ = 0
free = []
deliv = []
min_cost_f = 0
min_cost_d = 0
for i in vendors:
    a = i['delivery']['price']
    b = i['delivery']['minOrderPrice']
    all_ += 1
    if a['value'] != 0:
        deliv.append(i['name'])
        min_cost_d += int(b['value'])
    else:
        free.append(i['name'])
        min_cost_d += int(b['value'])
print(f'{all_} - всего, {len(free)/all_} - процент кафе с бесплатной доставкой, средняя минимальная цена - {min_cost_f/len(free)}, {len(deliv)/all_} - процент кафе с платной доставкой, средняя минимальная цена - {min_cost_d/len(deliv)}')
```

```python
print(all_, len(free), len(deliv))
```

```python
f_m = 0
f_c = 0
d_m = 0
d_c = 0
for i in vendors:
    a = i['delivery']['price']
    if a['value'] != 0:
        if i['categoryId'] == 1:
            d_c += 1
        else:
            d_m += 1
    else:
        if i['categoryId'] == 1:
            f_c += 1
        else:
            f_m += 1
print(f'Доля расторанов из первой категории (бесплатной доставкой) - {f_c/len(free)}, доля магазинов -   {f_m/len(free)}, доля расторанов из второй категории - {d_c/len(deliv)}, доля магазинов -  {d_m/len(deliv)}')
```
